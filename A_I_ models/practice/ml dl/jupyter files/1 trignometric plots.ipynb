```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import math as m

pi =m.pi

#linspace divides 0 to 10 range in 4 equal segments
print(np.linspace(0,10,5))

x= np.linspace(0,2*pi,30)
y= np.sin(x)

b = np.linspace(0,2*pi,30)
c = np.cos(b)

d = np.linspace(0,2*pi,30)
e = np.tan(d)


print(y)

print(plt.plot(x,y))
```

```python
z = np.linspace(0,4*pi,8)
a = np.sin(z)
#print(plt.plot(z,a))
print(plt.scatter(z,a,c="pink",marker="^",s = 200))

# or color ="pink"
```

```python
print(plt.plot(x,y,color="green",marker ="+"))
```

```python
print(plt.plot(z,a,c="yellow"))
print(plt.plot(x,y,color="green",marker ="+"))
plt.xticks(range(0,10))
plt.yticks([-1,0,1])
print(plt.plot(b,c,color="brown"))
```

```python
plt.figure(figsize=(20,9))
print(plt.plot(d,e,color="darkblue", label = "tan curve"))
print(plt.plot(z,a,c="yellow"))
print(plt.plot(x,y,color="green",marker ="+", label = "sin curve"))
print(plt.plot(b,c,color="brown",marker ="^", label ="cos curve"))
plt.legend()
plt.title(" sin cos tan graph" , fontsize = 20, color ="orange")
plt.xlabel("radian values", fontsize =15)
plt.ylabel("sin cos tan", fontsize =15)
plt.grid()
plt.show()
```
