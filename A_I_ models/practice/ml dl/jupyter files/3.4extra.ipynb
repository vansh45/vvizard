```python
import pandas as pd
import numpy as np
```

```python
path ='excelsample.xlsx'
df = pd.read_excel(path, header = None)
df
```

```python
df.dtypes
```


## Adding headers 

```python
headers =['maths', 'science','sst','hindi','english','city']
df.columns = headers
df
```

```python
nf = df.copy()
```

```python
names =['Ross','Rachel','Monika','Joey','Chandler']
nf['names'] = names
nf
```

```python
df.insert(0, "Name", names, True)
df
```

```python
mean = df['english'].mean()
```

```python
df.replace(np.NaN,mean,inplace =True)
```

```python
c =['vds','vidisha','v']
df.replace(c,'Vidisha',inplace =True)
df
```

```python
df.city = df['city'].str.capitalize()
```

```python
df["english"] = pd.to_numeric(df["english"])
df
```

```python
#df['english'] = pd.to_numeric(df['english'], downcast='integer')        
#or
df['english'] = df['english'].astype('int')
df
```

```python
avg=[0,0,0,0,0]
for j in range(0,5):
    a =0
    for i in range(1,6):
        a = a+ df.iloc[j][i]
    a= a/5
    avg[j]= a
    
avg
```

```python
P = avg.copy()
for aa in range(len(avg)):
    P[aa]= avg[aa]*2
P
```

```python

df['percentage'] = P
df
```

### Binning


```python
# IBM data analysis with python week2 binning 

#bins = np.linspace(min(df['percentage']),max(df['percentage']),6)
#for distributing between max and min already given in data
```

```python
bins = np.linspace(0,100,6)
# 6 equally spaced numbers or 5 divisions
grades =['F','D','C','B','A']
df['Grades'] = pd.cut(df['percentage'],bins,labels = grades,include_lowest =True)
```
