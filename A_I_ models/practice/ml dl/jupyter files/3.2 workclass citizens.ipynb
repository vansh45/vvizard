```python
# Reading XML file
import xml.etree.ElementTree as etree
tree = etree.parse("employee.xml")
root = tree.getroot()
columns = ["Name","Gender", "Emails"]
df = pd.DataFrame(columns = columns)
for node in root:
    name = node.find('name').text
    gender = node.find('gender').text
    email = node.find('email').text
    
    df = df.append(pd.Series([name, gender, email], index = columns),ignore_index =True)


df.head(5)
```
