```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
y_test = y_test.reshape(-1,)
```

```python
classes = ['airplane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

```python
def plot_sample(x,y,index):
    plt.figure(figsize = (15,2))
    plt.imshow(x[index])
    plt.xlabel(classes[y[index]])
```

```python
for i in range(0,10,2):
    plot_sample(x_train, y_train, i)
```

```python
x_train = x_train/255
```

```python
x_test = x_test/255
```

```python
cnn2 = models.Sequential([
    layers.Conv2D(32,(3,3),activation = 'relu', input_shape = (32, 32, 3)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64,(3,3),activation = 'relu'),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(64,activation ='relu'),
    layers.Dense(10,  activation ='softmax')
])

cnn2.compile(optimizer = 'adam',
           loss = 'sparse_categorical_crossentropy',
           metrics = ['accuracy'])

cnn2.fit(x_train, y_train, epochs = 25)
```

```python
cnn2.fit(x_test, y_test)
```

```python
cnn2.evaluate(x_test, y_test)
```

```python
import os

path = 'C://Users//Vansh Samaiya//Music//CIFAR10//test_Image//'

def ex(path):
    fig, axes = plt.subplots(4, 5, figsize=(20, 15))  # 4 rows, 5 columns

    for i in range(0, 21):  # Loop for 20 images
        p = os.path.join(path, 't{}.jpg'.format(i))

        img = image.load_img(p, target_size=(32, 32))
        img_array = image.img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0)
        img_array /= 255

        predictions = cnn2.predict(img_array)
        predicted_class = np.argmax(predictions[0])

        ax = axes[(i-1)//5, (i-1)%5]  # Determine the correct subplot for this image
        ax.imshow(img)
        ax.set_title(classes[predicted_class], fontsize=20)  # Adjust font size


        ax.axis('off')

        print(f'Predicted Class {i}:', classes[predicted_class])
        print(f'Confidence {i}:', predictions[0][predicted_class])

    plt.tight_layout()
    plt.show()

ex(path)
```
