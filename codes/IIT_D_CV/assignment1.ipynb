```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def threshold(l):
    res = 0
    for i in range(len(l)):
        for j in range(len(l[0])):
            if i !=1 or  j !=1:
                if l[i][j] < l[1][1]:
                    l[i][j] = 0
                else:
                    l[i][j] = 1
    return l
```

```python
l = [[5,12,3],[6,2,4],[8,3,7]]
print(threshold(l))
```

```python
def threshold_(img,centre,x,y):
  val = 0
  try:
    if img[x][y]>=centre:
      val = 1
  except:
      pass
  return val
```

```python
def lbp_pixel(img,x,y):
  lbp_arr = []
  lbp_arr.append(threshold_(img,img[x][y],x-1,y-1)) #top left
  lbp_arr.append(threshold_(img,img[x][y],x-1,y))   #top
  lbp_arr.append(threshold_(img,img[x][y],x-1,y+1)) #top right
  lbp_arr.append(threshold_(img,img[x][y],x,y-1))   #left
  lbp_arr.append(threshold_(img,img[x][y],x,y+1))   #right
  lbp_arr.append(threshold_(img,img[x][y],x+1,y-1)) #bottom left
  lbp_arr.append(threshold_(img,img[x][y],x+1,y))   #bottom
  lbp_arr.append(threshold_(img,img[x][y],x+1,y+1)) #bottom right

  val = 0
  for i in range(len(lbp_arr)):
    val+=lbp_arr[i]*(2**i)

  return val
```

```python
def LBP(img,x,y):
  img_lbp = np.zeros((x,y),np.uint8)
  for i in range(x):
    for j in range(y):
      img_lbp[i][j] =  lbp_pixel(img,i,j)
  return img_lbp
```
